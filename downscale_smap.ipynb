```python
from SFIM import SFIM
import rasterio 
import numpy as np
import pandas as pd
import os
```

```python
input_dir = 'dataset/VH1km'
sites = os.listdir(input_dir)
print(sites)
```

```python
ratio = 9
input_dir = 'dataset/VH1km'
sites = os.listdir(input_dir)
for site in sites:
    output_dir = f'dataset/smap1km_pm_ASC/{site}'
    os.makedirs(output_dir, exist_ok= True)

    site_path = f"{input_dir}/{site}"
    csv_path = f'daily_smap/{site}_smap.csv'

    df = pd.read_csv(csv_path)
    valid_dates = df.dropna(subset='smap_lon')
    dates = valid_dates['time'].tolist()

    for date in dates:
        file = f'{site_path}/{site}_S1_{date}_ASCENDING_VH1km.tif'
        output_path = f"{output_dir}/smap_pm_1km_{site}_{date}.tif"

        if not os.path.exists(file):
            print(f"File not found: {file}")
            continue
        with rasterio.open(file) as src:
            data = src.read(1)
            profile = src.profile 

        data_array = np.array(data)
        # ASCENDING corresponds to smap pm
        smap_value = valid_dates.loc[valid_dates['time'] == date, 'smap_soil_moisture_pm'].values[0]
        print(f"date : {date} ---- smap-pm : {smap_value}")
        smap_pixel = np.array([[smap_value]])*255.0

        smap_1km = SFIM(data_array, smap_pixel, ratio)
        normalized_smap_1km = smap_1km/255.0
        with rasterio.open(output_path, 'w', **profile) as dst:
            dst.write(normalized_smap_1km, 1)
            print(f"Downscaled : {date}")
```

```python
input_image = 'dataset/VH1km/SCAN_Kemole_Gulch/SCAN_Kemole_Gulch_S1_2020-01-06_ASCENDING_VH1km.tif'

ratio = 9 

smap = np.array([[0.3452]])*255

with rasterio.open(input_image) as src:
    data = src.read(1)
    profile = src.profile

data_array = np.array(data)

smap_1km = SFIM(data_array, smap, ratio)
smap_1km = smap_1km/255.0
with rasterio.open('dataset/smap_1km.tif', 'w', **profile) as dst:
    dst.write(smap_1km, 1)
smap_1km
```
